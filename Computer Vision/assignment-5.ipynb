```python
# perform histogram equalization and enhance contrast using histogram equalization
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = cv.imread('sample.jpg', cv.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Histogram for the Image')
plt.show()
```

```python
equ = cv.equalizeHist(img)
plt.hist(equ.ravel(), 256, [0, 256])
plt.title('Histogram for Equalized Image')
plt.show()
```

```python
plt.imshow(equ, cmap='gray')
plt.axis('off')
plt.title("Equalized Image")
plt.show()
```

```python
# side by side subplots
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title("Original Image")

plt.subplot(2, 2, 2)
plt.imshow(equ, cmap='gray')
plt.axis('off')
plt.title("Equalized Image")

plt.subplot(2, 2, 3)
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Histogram for Original Image')

plt.subplot(2, 2, 4)
plt.hist(equ.ravel(), 256, [0, 256])
plt.title('Histogram for Equalized Image')

plt.show()
```
