```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread("sample-1.jpg")
```

```python
img.shape
```

```python
# convert to RGB
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
print(f"Height: {img.shape[0]}")
print(f"Width: {img.shape[1]}")
print(f"Channels: {img.shape[2]}")
```

```python
# print data type of pixel
print(f"Data Type: {img.dtype}")
```

```python
print(f"Max intensity: {np.max(img)}")
print(f"Min intensity: {np.min(img)}")
```

```python
s_row = 230
e_row = 380
s_col = 300
e_col = 400
```

```python
# find the reigon of interest
img_roi = img[s_row:e_row, s_col:e_col]
plt.imshow(img_roi)
plt.axis('off')
plt.show()
```

```python
for i in range(s_row, e_row):
    for j in range(s_col, e_col):
      # find average of all 3 channels
      avg = np.mean(img[i, j])
      img[i, j] = (avg, avg, avg)
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
plt.imshow(img_roi)
plt.axis('off')
plt.show()
```

```python
h, w = 300, 500
image = np.zeros((h, w, 3), dtype=np.uint8)
image[:] = (50,50,50)
```

```python
lb = np.array([53,48,48])
ub = np.array([57,52,52])
mask = cv.inRange(image, lb, ub)
```
