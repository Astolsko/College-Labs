```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("sample-1.jpg")
img.shape # 3 color channels
```

```python
# Cv gets as BGR so need to convert to RGB
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
# convert to gray scale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gray.shape # no color channel
```

```python
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# thresholding
_, img_binary = cv.threshold(img_gray,135,255,cv.THRESH_BINARY)
img_binary.shape
```

```python
plt.imshow(img_binary, cmap = "gray")
plt.axis("off")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img_rgb)
axes[0].set_title("RGB Image")
axes[0].axis("off")

axes[1].imshow(img_gray, cmap="gray")
axes[1].set_title("Grayscale Image")
axes[1].axis("off")

axes[2].imshow(img_binary, cmap="gray")
axes[2].set_title("Binary Image")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
