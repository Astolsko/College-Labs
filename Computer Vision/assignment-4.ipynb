```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
```

```python
def log_transform(x):
    c = 255 / np.log(1 + np.max(x))
    return c * np.log(1 + x)

def power_transform(x, gamma):
    return np.power(x, gamma)

def negative_transform(x):
    max = np.max(x)
    return max - x
```

```python
# get an image 
img = cv.imread('sample.jpg')

img.shape
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gray.shape
```

```python
img_gray
```

```python
# display grayscale image
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# log transform
img_log = log_transform(img_gray)
```

```python
plt.imshow(img_log, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# power transform
power = [0.5,1,1.5,2]
for i in power:
    img_power = power_transform(img_gray, i)
    plt.imshow(img_power, cmap='gray')
    plt.axis('off')
    plt.title(f'Power Transform with gamma={i}')
    plt.show()
```

```python
img_negative = negative_transform(img_gray)
plt.imshow(img_negative, cmap='gray')
plt.axis('off')
plt.title('Negative Transform')
plt.show()
```
